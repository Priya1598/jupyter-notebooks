```python
n6= pd.read_csv('C:\data set/Position_Salaries.csv')
n6
```

```python
k=n6.iloc[:,1:2]
c=n6.iloc[:,2]
```

```python
k
```

```python
#fitting linear regression to the dataset
from sklearn.linear_model import LinearRegression
b=LinearRegression()
b.fit(k,c)
```

```python
salary=b.predict([[8]])
salary
```

```python
salary1=b.predict([[8.3]])
salary1
```

```python
#visulasing the  linear regression  results
```

```python
plt.scatter(k,c, color='red')
plt.plot(k,b.predict(k),color='blue')
plt.title('Linear Regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
# check session 10 at 1hr 12 min for more reference, detailed the problem statementin session 10
# linear regression doesnt suits the requirement so moving to polynomial regression to know the requirement check session
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree = 4)
k_poly=poly_reg.fit_transform(k) # only the independent variable need to be fit in fit_transform
lin_reg=LinearRegression()
lin_reg.fit(k_poly,c)
```

```python
#visulasing the polynomial regression result
```

```python
plt.scatter(k,c,color = 'red')
plt.plot(k,lin_reg.predict(k_poly),color="blue")
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.show()
```

```python
#predicting a result with poly regression
```

```python
s=lin_reg.predict(poly_reg.fit_transform([[8.3]]))
s
```
