```python
# remove all the useless column.
df = df.drop(['Over18'],axis=1)
df = df.drop(['EmployeeNumber'],axis=1)
df = df.drop(['StandardHours'],axis=1)
df = df.drop(['EmployeeCount'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
    if df[column].dtype == np.number:
        continue
    df[column] = LabelEncoder().fit_transform(df[column])
```

```python
x = df.drop(['Attrition'],axis=1)
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y,test_size=0.25, random_state=125)
```

```python
(x_train.shape),(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
mm = MinMaxScaler(feature_range=(0,1))
x_train_mm = mm.fit_transform(x_train)
x_test_mm = mm.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
accuracy_score(y_train,model.predict(x_train))
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
# under sampling
from collections import Counter
from imblearn.under_sampling import NearMiss
nm = NearMiss()
x_train_nm,y_train_nm = nm.fit_resample(x_train,y_train)
```

```python
Counter(y_train)
```

```python
Counter(y_train_nm)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier()
rc1 = rc.fit(x_train_nm,y_train_nm)
```

```python
confusion_matrix(y_test,rc1.predict(x_test))
```

```python
# over sampling
from imblearn.over_sampling import RandomOverSampler
rs = RandomOverSampler()
x_train_rs,y_train_rs = rs.fit_resample(x_train,y_train)
```

```python
Counter(y_train)
```

```python
Counter(y_train_rs)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier()
rc2 = rc.fit(x_train_rs,y_train_rs)
```

```python
confusion_matrix(y_test,rc2.predict(x_test))
```

```python
# combine (50% from your 1st class and 50% from 2nd class)
```

```python
from imblearn.combine import SMOTETomek
st = SMOTETomek()
x_train_st,y_train_st = st.fit_resample(x_train,y_train)
```

```python
Counter(y_train_st)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train_st,y_train_st)
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
confusion_matrix(y_train_st,model.predict(x_train_st))
```

```python
accuracy_score(y_test,model.predict(x_test))
```
