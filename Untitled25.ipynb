```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/akjadon/HH/master/Projects/heart_disease/heart.csv')
```

```python
f, (ax1, ax2) = plt.subplots(1,2,figsize =(15, 8))
corr = df.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
heatmapkws = dict(linewidths=0.1) 
sns.heatmap((df[df['target'] ==1]).corr(), vmax = .8, square=True, ax = ax2, cmap = 'YlGnBu', mask=mask, **heatmapkws);
ax1.set_title('Disease', fontsize=18)
sns.heatmap((df[df['target'] ==0]).corr(), vmax = .8, square=True, ax = ax1, cmap = 'afmhot', mask=mask,**heatmapkws);
ax2.set_title('Healthy', fontsize=18)
plt.show()
```

```python
x = df.drop(['target'],axis=1)
y = df['target']
```

```python
import statsmodels.formula.api as smf
model = smf.ols(formula='y ~ age+sex+cp+trestbps+chol+fbs+restecg+thalach+exang+oldpeak+slope+ca+thal',data=df).fit()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,stratify=y,random_state=10)
```

```python
(x_train.shape),(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
sc = RobustScaler()
model = sc.fit(x_train,y_train)
x_train = model.fit_transform(x_train)
x_test = model.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=100,verbose=True)
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
accuracy_score(y_test,lr.predict(x_test))
```

```python
accuracy_score(y_train,lr.predict(x_train))
```

```python
f,ax=plt.subplots(3,2,figsize=(12,12))
f.delaxes(ax[2,1])

for i,feature in enumerate(['age','thalach','chol','trestbps','oldpeak','age']):
    sns.distplot(df[df['target']==0][(feature)], ax=ax[i//2,i%2], kde_kws={"color":"white"}, hist=False )
    sns.distplot(df[df['target']==1][(feature)], ax=ax[i//2,i%2], kde_kws={"color":"white"}, hist=False )

    # Get the two lines from the ax[i//2,i%2]es to generate shading
    l1 = ax[i//2,i%2].lines[0]
    l2 = ax[i//2,i%2].lines[1]

    # Get the xy data from the lines so that we can shade
    x1 = l1.get_xydata()[:,0]
    y1 = l1.get_xydata()[:,1]
    x2 = l2.get_xydata()[:,0]
    y2 = l2.get_xydata()[:,1]
    ax[i//2,i%2].fill_between(x2,y2, color="darkorange", alpha=0.6)
    ax[i//2,i%2].fill_between(x1,y1, color="darkturquoise", alpha=0.6)

    #grid
    ax[i//2,i%2].grid(b=True, which='major', color='grey', linewidth=0.3)
    
    ax[i//2,i%2].set_title('{} by target \n (orange = disease)'.format(feature), fontsize=18)
    ax[i//2,i%2].set_ylabel('count', fontsize=12)
    ax[i//2,i%2].set_xlabel('Modality', fontsize=12)

    #sns.despine(ax[i//2,i%2]=ax[i//2,i%2], left=True)
    ax[i//2,i%2].set_ylabel("frequency", fontsize=12)
    ax[i//2,i%2].set_xlabel(str(feature), fontsize=12)

plt.tight_layout()
plt.show()
```
