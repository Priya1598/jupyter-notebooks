```python
df = df[df['total_rooms'] < 3000]
df = df[df['total_rooms'] > 1000]
```

```python
df = df[df['total_bedrooms'] < 650]
#df = df[df['total_bedrooms']<100]
```

```python
df  = df[df['population']<2500]
```

```python
df = df[df['households']<750]
```

```python
df = df[df['median_income']<7]
```

```python
df = df[df['median_house_value']<400000]
```

```python
df.shape
```

```python
x=df.drop(['median_house_value'],axis=1)
y=df['median_house_value'] #values is for array format
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
```

```python
x_train_scaled = sc.transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
pd.DataFrame(x_train_scaled)
```

```python
pd.DataFrame(x_test_scaled)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range=(0,1))
mms.fit(x_train)
x_train_mms = mms.transform(x_train)
x_test_mms = mms.transform(x_test)
```

```python
pd.DataFrame(x_test_mms)
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
rs.fit(x_train)
```

```python
x_train_rs = rs.transform(x_train)
x_test_rs = rs.transform(x_test)
```

```python
pd.DataFrame(x_train_rs)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from math import sqrt
```

```python
print(r2_score(y_test,lr.predict(x_test)))
```

```python
print(r2_score(y_train,lr.predict(x_train)))
```
